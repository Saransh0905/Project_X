```python
#subsequence with rk and rk 
import pandas as pd
from collections import Counter 


data = open("/home/saransh/Documents/Github/Project_X/organisedData/SignalP_output_blasted_starred.txt","r")
l = data.readlines()
seq = []
code = []
st = ''
for i in range(len(l)):
    if l[i][0]!='>':
        st = st+l[i][0:len(l[i])-1]
    else:
        code.append(l[i][0:len(l[i])-1])
        seq.append(st)
        st = ''
seq.append(st)
seq.remove('')

ans = []

numbering = []
count = -1
for s in seq:
    count+=1
    j = len(s)-1
    while j!=0:
        if s[j]+s[j-1] =='rk' or s[j]+s[j-1] =='kr' or s[j]+s[j-1] =='kk' or s[j]+s[j-1] =='rr':
            for k in range(j-1,0,-1):
                if s[k]+s[k-1] =='rk' or s[k]+s[k-1] =='kr' or s[k]+s[k-1] =='kk' or s[k]+s[k-1] =='rr':
                    ans.append(s[k-1:j+1])
                    numbering.append(count)
                    j = k+1
                    break
        j-=1
        
out = open("/home/saransh/Documents/Github/Project_X/organisedData/SignalP_starred_found_RK_&_RK_subsequence.txt","w")
cou  = 0
for i in range(len(seq)):
    #ind = 1
    if numbering.count(i)>0:
        cou +=1
        out.write(code[i])
        out.write("\n")
        for j in range(len(ans)):
            if numbering[j]==i:
                #out.write(str(ind)+". ")
                out.write(ans[j])
                out.write("\n")
                #ind+=1
```
